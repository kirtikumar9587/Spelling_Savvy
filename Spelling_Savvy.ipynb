# 3.TEXT PREPROCESSING

### Splitting

```python
def split(word):  
    parts = []
    for i in range(len(word) + 1):
        parts += [(word[ : i], word[i : ])]
    return parts
```

Now after this we can add ,delete

### 3.1 DELETE

### 3.1) Delete

'loave' -> 'love'

```python
#here to delete one character at a time os for that 1 remove l then o upto e
#take r after one index
def delete(word):
    
    output = []
    for l,r in split(word):
        output.append(l + r[1:])
    return output

delete("loave")#this is get the correct word love
```

### 3.2) Swap

'lvoe' -> 'love'


```python
def swap(word):
        
    output = []    
    for l,r in split(word):
        if (len(r) > 1):
            output.append(l + r[1] + r[0] + r[2:])
    return output
            
swap('lvoe')
```

## 4. Finding the Prediction (Level - 1)
### 4.1) Combining Possible Words

we take a word and pass through all above 4 operations and then combine all value in list and then ofund unique word in it

```python
def edit(word):   
    return list(set(insert(word) + delete(word) + swap(word) + replace(word)))
```

### 4.2) Predicting the Word

```python
def spell_check_edit_1(word, count = 5):
    
    output=[]
    suggested_words = edit(word)
    for wrd in suggested_words:        
        if wrd in word_probability.keys():
            output.append([wrd, word_probability[wrd]])
            
    return list(pd.DataFrame(output, columns = ['word','prob']).sort_values(by = 'prob', ascending = False).head(count)['word'].values)
```

## 5. Finding the Prediction (Level - 2)
### 5.1) Combining Possible Words

```python
def spell_check_edit_2(word, count = 5):
    # Define an empty list to store the final output words
    output = []
    # Get the first level edits for the given word
    suggested_words = edit(word)       # Level one Edit
     # Loop through each word from the first level edits
    for e1 in edit(word):
        # Add second level edits to the suggested_words list
        suggested_words += edit(e1)    # Second Level Edit 
    
    # Remove duplicates by converting the list to a set and back to a list
    suggested_words = list(set(suggested_words))
    
    # Create a DataFrame from the output list, sort it by the 'prob' column in descending order,
    # select the top 'count' words, and extract their values as a list
    for wrd in suggested_words:
        if wrd in word_probability.keys():
            output.append([wrd, word_probability[wrd]])
    return list(pd.DataFrame(output, columns = ['word','prob']).sort_values(by = 'prob', ascending = False).head(count)['word'].values)
        
spell_check_edit_2('fameli')
```
